```python
import nltk
import os
import pandas as pd
from nltk.text import Text
from matplotlib import pyplot as plt
from nltk import regexp_tokenize
import string
from collections import Counter
from nltk import ngrams
from nltk import word_tokenize
from collections import Counter
from nltk.probability import FreqDist
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer,PorterStemmer,SnowballStemmer
from nltk.corpus import stopwords
import re
import glob
import errno
```

```python
# load text  
with open('/users/raissasoares/desktop/thesis/ny_sun/pre_processed_1897/1899_1_8.txt', 'r', encoding='utf-8') as f:
        text = f.read()
        print (text[10:])
```

```python
# split based on words only
words = re.split(r'\W+', text)
text = re.sub(r'[?|!|\'|"|#|K|¢|*|)|(|:|-|_|__|»|«|<|>|®|$|°|-~|;|—|~|}|{|&|———]|\+|=+|-‘|==|\\|©|§|k]',r'',text)
text = re.sub("[^a-zA-Z]", " ", text)
text = re.sub(r'[.|,|)|(|\|/]',r' ',text) # remove newline and spaces
text = text.replace('\n', '')
text_clean = re.sub('\s+', ' ', text).strip()
print(words[:100])
```

```python
# split into sentences
from nltk import sent_tokenize
sentences = sent_tokenize(text_clean)
print(sentences[:10])
```

```python
# split into words, no punctuation, only isalpha, lowercase
tokens = word_tokenize(text_clean)
tokens = [i for i in tokens if i not in string.punctuation]
tokens = [word.lower() for word in words]
tokens = [word for word in tokens if word.isalpha()]
print(tokens[:100])
```

```python
stop = set(stopwords.words('english'))
stoptoken = [] 
for i in tokens:
    if i not in stop:
        stoptoken.append(i)
print(stop)
print(stoptoken)
with open('1899_1_8.txt', 'w', encoding='utf-8', errors='ignore') as f:
        f.write(str(stoptoken))
```

```python
# stemming of words
stemmer = SnowballStemmer("english")
stemmed = [stemmer.stem(words) for words in stoptoken]
print(stemmed[:100])
```

```python
print(Counter(stoptoken).most_common(30))
```

```python
bigrams = list(ngrams(stoptoken, 2))
freq_bigrams = nltk.FreqDist(bigrams)
print(freq_bigrams.most_common(10))
```
